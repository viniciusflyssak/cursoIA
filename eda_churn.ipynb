## Análise Bivariada

```python
# Hipótese: Clientes com contrato tipo mensal com maior probabilidade de Churn
# Construir tabela de contingência
pd.crosstab(df_churn.Churn, df_churn.Contract, margins=True, margins_name='Total')
```

```python
# Hipótese: Clientes com contrato tipo mensal com maior probabilidade de Churn - proporção
# Construir tabela de contingência
pd.crosstab(df_churn.Churn, df_churn.Contract, margins=True, margins_name='Total', normalize='index')
```

Considerando o público total, os clientes de contrato mensal representam 55%. Porém, se considerarmos apenas os clientes que abandonaram o serviço, 88% tinham contrato mensal. Dessa forma, podemos afirmar que há uma correlação entre o tipo de contrato mensal e o abandono.

```python
# Avaliar a correlação entre duas variáveis categoricas (qualitativas)

# Executar teste de hipótese chamado Chi-Square ou Qui-Quadrado de Pearson
# Num teste de hipótese, duas hipóteses são formuladas:
# H0 (hipótese nula): As variáveis são independentes
# H1 (hipótese alternativa): As variáveis são dependentes

# o teste serve para confirmar ou recusar a hipótese nula (H0)
# Quando a probalidade de observarmos H0 é inferior a 0.05 (p-value),
# recusamos H0 e seguimos com a complementar
```

```python
# Gerar um DF da CrossTab (sem totais)
df_crosstab_churn_contract = pd.crosstab(df_churn.Churn, df_churn.Contract)
```

```python
# Calcular os Scores e P-Value
from scipy.stats import chi2_contingency
```

```python
# Tupla = lista imutável
chi_scores_churn_contract = chi2_contingency(df_crosstab_churn_contract)
```

```python
chi_scores_churn_contract  
```

```python
scores_churn_content = pd.Series(chi_scores_churn_contract[0])
```

```python
pvalues_churn_content = pd.Series(chi_scores_churn_contract[1])
```

```python
# Apresentamos números com decimais sem notação científica
pd.set_option('display.float_format', lambda x: '%.15f' % x)
```

```python
# Criar dataframe com os resultados
df_chi_churn_contract = pd.DataFrame({'Qui2': scores_churn_content, 'P-Value': pvalues_churn_content})
```

Conforme teste estatístico (Chi-Square), o P-value <= 0.05, desta forma rejeitamos a hipótese nula, ou seja, as variáveis não são independentes. Pelo Qui2 alto, podemos afirmar que há uma forte correlação.

```python
# Avaliar a correlação entre uma variável qualitativa e uma quantitativa
# Usar o mesmo teste de Chi-Square, mas transformar a variável quantitativa em qualitativa
# Hipótese: Clientes com menos de 6 meses de contrato tem maior probabilidade de Churn

# Criar uma coluna nova no Dataframe com base na condição de outra coluna
# Coluna = 'TempoMenor6Meses'

import numpy as np
```

```python
# Criar coluna nova
df_churn['TempoMenor6Meses'] = np.where(df_churn.tenure < 6, 'Yes', 'No')
```

```python
# Checar o dataframe
df_churn.head(5)
```

```python
# Gerar um DF da CrossTab (sem totais)
df_crosstab_churn_tenure = pd.crosstab(df_churn.Churn, df_churn.TempoMenor6Meses)
```

```python
# Calcular os Scores
chi_scores_churn_tenure = chi2_contingency(df_crosstab_churn_tenure)
```

```python
scores_churn_tenure = pd.Series(chi_scores_churn_tenure[0])
pvalues_churn_tenure = pd.Series(chi_scores_churn_tenure[1])
```

```python
# Criar dataframe com os resultados
df_chi_scores_churn_tenure = pd.DataFrame({'Qui2': scores_churn_tenure, 'P-Value': pvalues_churn_tenure})
```

```python
# Analisar os resultados
df_chi_scores_churn_tenure
```

Como o p-value é menor que 0.05, rejeitamos a hipótese nula (H0) e aceitamos a hipótese alternativa (H1)
ou seja, as variáveis não são independentes. A correlação entre churn e tempo de contrato menor que 6 meses
é menor que a correlação entre churn e tipo de contrato

```python
# Correlação entre duas variáveis numéricas
# Tenure com total charges
# A intuição é que quanto mais tempo de contrato, maior o valor pago.

# Correlação entre 2 variáveis numéricas - Pearson
df_churn.tenure.corr(df_churn.TotalCharges)
```

```python
# Correlação entre 2 variáveis numéricas - Spearman
df_churn.tenure.corr(df_churn.TotalCharges, method='spearman')
```

Há uma forte correlação entre tenure e total charges, usando métodos estatísticos para medir a correlação
